Comenzaré comprobando la frecuencia de abandonos y no abandonos en los datos de entrenamiento.

```python
perdidos = df_train['abandono'].isna().sum()
print("Hay ",perdidos, " valores perdidos en 'abandono'")

x= df_train['abandono'].value_counts()

print("Numero de 0s (No abandonos): ",x.get(0,0))
print("Numero de 1s (Abandonos): ",x.get(1,0))
```

Ahora lo mismo pero más visual.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='abandono',data=df_train)
plt.title('Distribución de "abandono"')
plt.xlabel('Abandono')
plt.show()
```

Después de comprobar que los abandonos suponen más o menos un cuarto del total; procederé comprobando la relación que existe entre el resto de variables por la objetivo. Comenzaré con aquellas varibles que no necesiten ningún tipo de transformación para comprobar esto.

### Variable 'trabaja'


```python
perdidos = df_train['trabaja'].isna().sum()
print("Hay ",perdidos, " valores perdidos en 'trabaja'")

x = pd.crosstab(df_train['trabaja'], df_train['abandono'])
x.plot(kind='bar', stacked=True, color=['skyblue', 'red'])
plt.title('Abandono según si trabaja o no')
plt.xlabel('Trabaja') 
plt.ylabel('Número de casos')
plt.show()

#Ahora lo mismo, pero normalizada, para ver mejor la relación de abandonos y no abandonos

x = pd.crosstab(df_train['trabaja'], df_train['abandono'], normalize='index') * 100
plt.figure(figsize=(8, 6))
x.plot(kind='bar', stacked=True, color=['skyblue', 'red'])
plt.title('Abandono según si trabaja o no (Normalizado)')
plt.xlabel('Trabaja') 
plt.ylabel('% de casos')
plt.show()
```

Parece que puede ser relevante, ya que si que existe diferencia notable en la proporción de abandonos entre la gente que trabaja y la que no.

### Variable 'horas_trabajo'

```python
perdidos = df_train['horas_trabajo'].isna().sum()
x = (df_train['trabaja']==0).sum()
print("Hay ",x, " valores 0 en trabaja")
print("Hay ",perdidos, " valores perdidos en 'horas_trabajo'")
y = df_train[df_train['trabaja'] == 1]['horas_trabajo'].isna().sum()
print("Hay ",y, " valores perdidos dentro de la gente que si ha trabajado")
```

Esto me dice que el resto de valores perdidos, realmente son 0s. Como esta variable tiene valores perdidos, voy a realizar una pequeña transformación, para comprobar si esos valores perdidos pueden aportar algo de información. Para no modificar el dataframe original, voy a realizar una copia para realizar la transformación en ella.

```python
df_copia =df_train.copy(deep=True)

# En las personas que no han trabajado, sustituyo los valores perdidos por un 0
df_copia.loc[df_copia['trabaja'] == 0, 'horas_trabajo'] = df_copia.loc[df_copia['trabaja'] == 0, 'horas_trabajo'].fillna(0)

# En las personas que si han trabajado les voy a poner los valores perdidos a -1 (solo para ver la distribución de estos)
df_copia.loc[df_copia['trabaja'] == 1, 'horas_trabajo'] = df_copia.loc[df_copia['trabaja'] == 1, 'horas_trabajo'].fillna(-1)

x = pd.crosstab(df_copia['horas_trabajo'], df_copia['abandono'])
x.plot(kind='bar', stacked=True, color=['skyblue', 'red'])
plt.title('Abandono según las horas que trabaja')
plt.xlabel('Horas que trabja')
plt.ylabel("Número de casos")
plt.show()

# Ahora normalizada
x = pd.crosstab(df_copia['horas_trabajo'], df_train['abandono'], normalize='index') * 100
x.plot(kind='bar', stacked=True, color=['skyblue', 'red'])
plt.title('Abandono según las horas que trabaja (Normalizado)')
plt.xlabel('Horas que trabja')
plt.ylabel("% de casos")
plt.show()
```

No parece aportar mucha más información que la anterior, de hecho una vez puesta de esta manera, practicamente es la misma pero con más información (si no fuera por los valores a -1 que son los verdaderamente perdidos).

### Variable 'modalidad'

## 2. Prepocesamiento
Ahora voy a pasar con el preprocesamiento, el cual lo llevaré a cabo utilizando una pipeline de scikit learn (como se pide). Las transfomaciones de datos que haré, serán las mismas que he realizado durante la fase del análisis exploratorio.

Antes que nada voy a importar todo lo necesario (aunque algunas cosas ya lo estuvieran antes).

```python
import pandas as pd
import numpy as np
import re
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

Ahora, lo primero será separar el datframe df_train en X e y

```python
# Primero cargo de nuevo el .csv. Aunque no lo haya modificado, de esta manera me aseguro de que es correcto.
df_train = pd.read_csv('uclm_student_train.csv')

# Separo el dataframe
y=df_train['abandono']
X=df_train.drop(columns=['abandono'])
```

Lo siguiente es declarar los tipos de variable segun quiero transformarlos.

```python
# Numéricas genéricas
numericas=[
    'nota_acceso', 'creditos_a1', 'superados_a1', 'nota_s1', 'satisfaccion',
    'posts_foro', 'uso_biblioteca', 'eventos', 'tutorias',
    'meses_matriculado', 'talla_zapato', 'horas_trabajo'
]

# Categóricas genéricas
categoricas=[
    'trabaja', 'modalidad', 'comentarios'
]

# Variables que necesitan una transformación específica
fecha =['nacimiento']
moodle = ['horas_moodle']
provincia = ['provincia']
bachillerato = ['bachillerato']

# El resto de variables, no las usaré, y las eliminaré en la pipeline. Son id, nombre, residencia_id, numero_fav, color_fav y grupo_trabajo
```

Ahora voy a crear mis transformadores específicos.

```python
# Transformador para nacimento
class TransformadorNacimiento(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X, y=None):
        X_dt = pd.to_datetime(X.iloc[:, 0], format='%Y-%m-%d', errors='coerce')
        return X_dt.apply(lambda fecha: fecha.year if pd.notna(fecha) else np.nan).to_frame()
    
# Transformador para horas_moodle
class TransformadorHorasMoodle(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X, y=None):
        def limpiar_horas_moodle(valor):
            match = re.search(r'[\d\.]+', str(valor))
            return float(match.group(0)) if match else np.nan
        return X.iloc[:, 0].apply(limpiar_horas_moodle).to_frame()
    
# Transformador para provincia
class TransformadorProvincia(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X, y=None):
        provincias_principales = {'Albacete', 'Cuenca', 'Ciudad Real', 'Toledo'}
        def limpiar_provincia(prov):
            return prov if prov in provincias_principales else 'Otro'
        return X.iloc[:, 0].astype(str).apply(limpiar_provincia).to_frame()
    
# Transformador para bachillerato
class TransformadorBachillerato(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X, y=None):
        diccionario = {
            'científico': 'ciencias', 'ciencias puras': 'ciencias',
            'cientifico-tecnologico': 'tecnologico', 'tecnológico': 'tecnologico',
            'tecnológico industrial': 'tecnologico', 'letras': 'humanidades',
            'económico': 'ciencias sociales', 'ccss': 'ciencias sociales',
            'ciencias': 'ciencias', 'humanidades': 'humanidades',
            'ciencias sociales': 'ciencias sociales', 'artes': 'artes'
        }
        serie = X.iloc[:, 0].astype(str).str.lower().str.strip().replace(diccionario)
        return serie.fillna('missing').to_frame()
```

Una vez tengo creados los transformadores, lo siguiente es definir las pipelines de transformación.

```python
transformador_numerico = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

transformador_categorico = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformador_moodle = Pipeline(steps=[
    ('cleaner', TransformadorHorasMoodle()),
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

transformador_provincia = Pipeline(steps=[
    ('grouper', TransformadorProvincia()),
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformador_bachillerato = Pipeline(steps=[
    ('mapper', TransformadorBachillerato()),
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformador_nacimiento = Pipeline(steps=[
    ('date_to_year', TransformadorNacimiento()),
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

Ahora los uno en el ColumnTransformer.

```python
preprocesador = ColumnTransformer(
    transformers=[
        ('moodle_pipe', transformador_moodle, moodle),
        ('date_pipe', transformador_nacimiento, fecha),
        ('prov_pipe', transformador_provincia, provincia),
        ('bach_pipe', transformador_bachillerato, bachillerato),
        ('num_pipe', transformador_numerico, numericas),
        ('cat_pipe', transformador_categorico, categoricas),
    ],
    remainder='drop' # Con esta linea, elimino todas las variables que no he transformado (id, nombre, residencia_id, numero_fav, color_fav y grupo_trabajo)
)
```

Ahora procedo con la creación de los pipelines finales. Serán dos, uno para el modelo Arbol de clasificación y el otro para Naive Bayes.

```python
# Para el Arbol de decisiones
pipeline_arbol = Pipeline(steps=[
    ('preprocessor', preprocesador),
    ('classifier', DecisionTreeClassifier(random_state=0)) 
])

# Para Naive Bayes
pipeline_naive_bayes = Pipeline(steps=[
    ('preprocessor', preprocesador),
    ('classifier', GaussianNB())
])
```

## 3. Validación

Para la validación cruzada, usaré la clase que viene de base implementada en scikit learn, cross_val_score; en concreto haré la 5-Fold. Las métricas que usaré son accuracy y f1.

```python
metricas=['accuracy', 'f1']

# Validación del Arbol de decisiones
print('Evaluación de Arbol de decisiones')

for metrica in metricas:
    # 5-Fold Cross-Validation para la métrica actual
    scores_arbol = cross_val_score(
        estimator=pipeline_arbol,
        X=X,
        y=y,
        cv=5,
        scoring=metrica,
        n_jobs=-1
    )
    media_arbol=scores_arbol.mean()
    desviacion_arbol=scores_arbol.std()

    print(metrica,'->  promedio: ',media_arbol,' ± ',desviacion_arbol)

print("\nEvaluación de Naive Bayes")
# Validación de Naive Bayes (Gausiano)
for metrica in metricas:
    scores_naive = cross_val_score(
        estimator=pipeline_naive_bayes,
        X=X,
        y=y,
        cv=5,
        scoring=metrica, 
        n_jobs=-1
    )
    media_naive=scores_naive.mean()
    desviacion_naive=scores_naive.std()

    print(metrica,'->  promedio: ',media_naive,' ± ',desviacion_naive)
```

## 4. Prediccion

En este apartado, utilizaré los modelos con las pipelines integradas que acabo de crear, para predecir la columna abandono del archivo uclm_student_test.csv; para posteriormente subir el resultado a kaggle.

```python
# Vuelvo a cargar df_test, por si acaso
df_test=pd.read_csv('uclm_student_test.csv')
```

Lo siguiente que haré será generar una función que pasada una pipeline y la variable que contiene el dataframe, me genere un fichero .csv con solo el campo 'id' y el compo 'abandono' predecido.

```python
def generar_archivo_csv(pipeline, df, nombre_archivo):

    # Guardo la columna par luego insertarla en el archivo de entrega (ya que lo elimino en la pipeline)
    ids = df['id'].copy() 

    # Predigo la columna 'abandono'
    predicciones_abandono = pipeline.predict(df)
    
    # Construyo el dataframe con los resultados y el id
    df_submission = pd.DataFrame({
        'id': ids,
        'abandono': predicciones_abandono
    })
    
    # Me aseguro de que el tipo de dato generado para la predicción de 'abandono' es int
    df_submission['abandono'] = df_submission['abandono'].astype(int)

    # Genero el csv
    df_submission.to_csv(nombre_archivo, index=False)

    return df_submission
```

Lo siguiente, y último, es generar los .csv con los resultados predecidos, en este caso crearé dos, uno para cada modelo implementado.

```python
# Creación del fichero con las predicciones del modelo Arbol de decisiones
nombre_arbol='arbol_kaggle.csv'

pipeline_arbol.fit(X,y)
generar_archivo_csv(pipeline_arbol, df_test, nombre_arbol)
```

```python
# Creación del fichero con las predicciones del modelo Naive Bayes (Gaussiano)
nombre_naive='naive_bayes_kaggle.csv'

pipeline_naive_bayes.fit(X,y)
generar_archivo_csv(pipeline_naive_bayes, df_test, nombre_naive)
```
